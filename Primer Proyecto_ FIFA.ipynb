```python
#Bucle para observar las filas que contienen "Free y On loan" mediante un bucle de ejecucion
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract']  or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
#Funcion para extraer informacion sobre el contrato
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan #nan not a number in numpy
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
    return start_date, end_date, contract_length

#Aplico la funcion para poder crear las 3 columnas de contrato y crear nuevas columnas
new_cols = ['Contract Start', 'Contract End', 'Contract Length(Years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

#Creo un bucle de ejecucion que se repetira a lo largo de todas las columnas del dataframe
for i in range(len(new_cols)):
    #insertamos las 3 columnas nuevas
    #en la localizacion correspondiente dentro del marco de datos
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column = new_cols[i], value= new_data[i])
```

```python
#Contract categories
def categorize_contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'
#Agrego la columna de estado del contrato
fifa.insert(fifa.columns.get_loc('Contract Length(Years)')+1, 'Contract Status', fifa['Contract'].apply(categorize_contract_status))
fifa.head()
```

```python
def convert_h(height):
    if 'cm' in height:
        return int(height.strip('cm')) #el metodo strip() elimina los caracteres que le son indicados
                                       # en forma arbitraria solo elimina los espacios en blancos que esten en el string
    else:
        feet, inches = height.split("'") #split transforma la cadena a una lista de valores
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches *2.54) #round para redondear el punto flotante de los cm
#aplico funcion altura a la columna Height
fifa['Height'] = fifa['Height'].apply(convert_h)
fifa.sample(20)
```

```python
#Cambio de nombre de la columna, metodo rename
fifa = fifa.rename(columns={'Height':'Height (cm)'})
```

```python
def convert_w(weight):
    if 'kg' in weight:
        return int(weight.strip('kg')) #el metodo strip() elimina los caracteres que le son indicados
                                       # en forma arbitraria solo elimina los espacios en blancos que esten en el string
    else:
        #convierto las libras a kilogramos
        pounds = int(weight.strip('lbs'))
        return round(pounds / 2.2046) #round para redondear el punto flotante de los cm
#aplico funcion altura a la columna Height
fifa['Weight (Kg)'] = fifa['Weight (Kg)'].apply(convert_w)
```

```python
#Cambio de nombre de la columna, metodo rename
fifa = fifa.rename(columns={'Weight':'Weight (Kg)'})
```
