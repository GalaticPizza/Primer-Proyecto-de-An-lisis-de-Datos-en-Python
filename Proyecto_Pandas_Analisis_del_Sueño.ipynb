## Analisis de edades

```python
df['Gender'].value_counts()
```

```python
df.Gender.value_counts()
```

```python
#observar sin tantos ceros los datos
pd.set_option('display.precision', 2)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df.describe()
```

```python
df['Age'].mean()
```

```python
df['Age'].std( )
```

problema con las edades del dataframe

```python
# hago referencia a una columna por el nombre de la columna
# genero una nueva columna y le adhiero un valor a la misma
# genero una condicional para separar y diferenciar adultos de adultos mayores respecto a la edad de los mismos
# siendo asi menor a 40 adulto y mayor a 41 adulto mayor
df.loc[df['Age'] <= 40, 'Age_group'] = 'Adult'
df.loc[df['Age'] >= 41 , 'Age_group'] = 'Older Adult'
df.head()
```

```python
df['Age_group'].value_counts()
```

existe gente mayor a 40 que gente de menor edad a 40, esto quizas se deba a que no hay un margen tan elevado en las edades, siendo esto un Outlier de este conjunto de datos

```python
  # Agrupo segun el genero las edades de los encuestados, y con el metodo agg, de agregate genero un dataframe que contiene
  # la media y la desv. estandar de las edades
df.groupby('Gender')['Age'].agg(['mean','std'])
```

obtengo informacion de las edades distinguiendo entre hombres y mujeres, siendo en este caso hay mujeres mucho mas mayores que de hombres

## Analisis de las ocupaciones

```python
df['Occupation'].unique()
```

```python
#cuento cuantas personas hay en cada profesion segun el genero
df.groupby('Occupation')['Gender'].value_counts()
```

```python
print(df['Sleep Duration'].min(),df['Sleep Duration'].max())
```

```python
df.groupby('Gender')[['Sleep Duration', 'Quality of Sleep']].mean()
```

```python
#Calidad y duracion del sueño en las distintas profesiones
ocup = df.groupby('Occupation')[['Sleep Duration', 'Quality of Sleep']].mean()
ocup.sort_values(by = 'Quality of Sleep')
```

```python
#El indice de masa Corp. por ocupaciones
bmi = df.groupby('Occupation')['BMI Category'].value_counts()
bmi.sort_values()
```

```python
steep = df.groupby('Occupation')['Daily Steps', 'Physical Activity Level', 'Sleep Duration', 'Quality of Sleep', 'Stress Level'].mean()
steep.sort_values(by= 'Quality of Sleep')
```
